```python
taille = len(donnees.Date)
minute = np.zeros((taille,1))
heure = np.zeros((taille,1))
jour = np.zeros((taille,1))
mois = np.zeros((taille,1))
annee = np.zeros((taille,1))
minute[0]=0;
heure[0]=0;
jour[0]=15;
mois[0]=4;
annee[0]=2015;
i=1
while i < taille :
    if minute[i-1] == 50 :
        heure[i]=heure[i-1]+1
        minute[i]=0
        mois[i]=mois[i-1]
        annee[i] = annee[i-1]
        if heure[i-1] == 23 :
            heure[i]= 0
            jour[i] = jour[i-1]+1
            if (mois[i-1] == 1) | (mois[i-1] == 3) | (mois[i-1] == 7) | (mois[i-1] == 8) | (mois[i-1] == 10) | (mois[i-1] == 12) :
                if jour[i-1]==31 :
                    jour[i]=1
                    if mois[i-1]==12 :
                        mois[i]=1
                        annee[i]=annee[i-1]+1
                    else :
                        mois[i]=mois[i-1]+1
                        annee[i]=annee[i-1]      
                else :
                    jour[i]=jour[i-1]+1
                    mois[i]=mois[i-1]
                
            elif (mois[i-1]==2) :
                if jour[i-1]==29 :
                    jour[i]=1
                    mois[i]=mois[i-1]+1
                    annee[i]=annee[i-1]
                else :
                    jour[i]=jour[i-1]+1
                    mois[i]=mois[i-1]
                    annee[i] = annee[i-1]
            elif (mois[i-1]==5) :
                if (annee[i-1]==2015) :
                    if jour[i-1]==30 :
                        jour[i]=1
                        mois[i]=mois[i-1]+1
                        annee[i]=annee[i-1]
                    else :
                        jour[i]=jour[i-1]+1
                        mois[i]=mois[i-1]
                        annee[i] = annee[i-1]
                else :
                    if jour[i-1]==31 :
                        jour[i]=1
                        mois[i]=mois[i-1]+1
                        annee[i]=annee[i-1]
                    else :
                        jour[i]=jour[i-1]+1
                        mois[i]=mois[i-1]
                        annee[i] = annee[i-1]
            else :
                if jour[i-1]==30:
                    jour[i]=1
                    mois[i]=mois[i-1]+1
                    annee[i]=annee[i-1]     
                else :
                    jour[i]=jour[i-1]+1
                    mois[i]=mois[i-1]   
        else :
            heure[i]=heure[i-1]+1
            jour[i]=jour[i-1]
            mois[i]=mois[i-1]
            annee[i] = annee[i-1]
    else :
        minute[i]=minute[i-1]+10;
        heure[i]=heure[i-1]
        jour[i] = jour[i-1]
        mois[i] = mois[i-1]
        annee[i] = annee[i-1]
    i=i+1

donnees["Year"] = annee
donnees["Month"] = mois
donnees["Day"] = jour
donnees["Hour"] = heure
donnees["Minute"] = minute





```
