```python
from mpl_toolkits.axes_grid1.parasite_axes import host_subplot
from mpl_toolkits.axisartist.axislines import Axes
from scipy.sparse.csr import csr_matrix
from sklearn import datasets
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error
from sklearn.svm.classes import NuSVR
from sklearn.ensemble.gradient_boosting import GradientBoostingRegressor
from sklearn.linear_model.stochastic_gradient import SGDClassifier
from sklearn.metrics import hamming_loss
import locale, time
# locale.setlocale(locale.LC_TIME, "fr_FR".encode('ascii'))
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# from matplotlib import rc
# from __future__ import unicode_literals
import numpy as np
import matplotlib
# matplotlib.rcParams['text.usetex'] = True
# matplotlib.rcParams['text.latex.unicode'] = True
# rc('font',**{'family':'serif','serif':['Times']})

%matplotlib nbagg


rhoair = 1.2
Cp = 16/27
Rayon_Palles = 101
pi=3.1415
puissCoeff = 0.5*Cp*pi*rhoair*Rayon_Palles*Rayon_Palles
Nseries = 5

donnees = pd.read_csv('Results_Used_QN2_Working_04-2015_04-2017.csv',sep=";")
donnees['Date'] = pd.to_datetime(donnees['Date'],infer_datetime_format=True)
donnees = donnees.set_index(['Date'])
df = [pd.DataFrame() for i in range(Nseries)]
for i in range(Nseries):
    df[i]["Puissance_Active"] = donnees["Active_Power_QN2" + str(i+1) + "-mean"]
    df[i]["Vitesse_Vent"] = donnees["Wind_Speed_QN2" + str(i+1) + "-mean"]
    df[i]["Puissance_Vent"] = puissCoeff*df[i].Vitesse_Vent**3

    # mise Ã  l'Ã©chelle de la puissance active
    mask = df[i]["Puissance_Active"] > 10000
    df[i]["Puissance_Active"][mask] /= 1000

    df[i]["Rendement"] = df[i].Puissance_Active*1000/df[i].Puissance_Vent
    mask1 = df[i]["Puissance_Active"] < 0
    df[i]["Rendement"][mask1] = 0
    mask2 = df[i]["Rendement"] > 1.00
    df[i]["Rendement"][mask2] /= 10 

plt.plot(df[4]["Vitesse_Vent"],df[4]["Rendement"])
```
